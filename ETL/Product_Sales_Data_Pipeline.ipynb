### Function to connect to AWS and download the files

```python
def downloadFromAws(filename, index):
    s3c = boto3.client('s3',aws_access_key_id=config.aws_access_key,
    aws_secret_access_key=config.aws_secret_key)
    remote_file_path = 'Data/'+filename
    obj = s3c.get_object(Bucket= config.aws_s3_bucket , Key = remote_file_path)
    if index == 0:
        df = pd.read_csv(io.BytesIO(obj['Body'].read()), encoding='utf8', index_col=0)
    else:
        df = pd.read_csv(io.BytesIO(obj['Body'].read()), encoding='utf8')
    return df
    
```

## Write the cleaned data into PostgreSQL database

```python
from sqlalchemy import create_engine
import psycopg2
import config
import pandas as pd
```

```python
db_string = f"postgresql://{config.db_user}:{config.db_password}@{config.db_hostname}/{config.db_name}"
engine = create_engine(db_string)
```

```python
# Function to write DataFrame to Postgress SQL database tables.
# Input params:
# dataframe -> dataframe to written
# table_name -> target table name
# conn -> conection engine
# action-> append or replace
# indexBool-> index column to be written or not.

def write_to_table(dataframe, table_name, conn, action, indexBool):
    rows_written=dataframe.to_sql(name=table_name,con=conn,if_exists=action, index=indexBool)
    return rows_written
```

```python
# product sheet data insert
rows = write_to_table(product_data_df,'products_sheet_processed',engine,'append',True)
print('Number of rows inserted : ', rows)

```

```python
# sales_team_sheet_processed  data insert
rows=write_to_table(sales_team_df,'sales_team_sheet_processed',engine,'append', True)
print('Number of rows inserted : ', rows)
```

```python
# customer_data_processed  data insert
rows=write_to_table(customer_data_df,'customer_data_processed',engine,'append', True)
print('Number of rows inserted : ', rows)
```

```python
# customer_data_processed  data insert
rows=write_to_table(store_region_data_df,'store_region_processed',engine,'append', False)
print('Number of rows inserted : ', rows)
```

```python
# sales_order_sheet_processed  data insert
rows=write_to_table(sales_order_df,'sales_order_sheet_processed',engine,'append',True)
print('Number of rows inserted : ', rows)
```
